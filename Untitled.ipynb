```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestRegressor

data=pd.read_excel('clean_data/.xlsx')
corr_matrix = data.corr()
plt.figure(figsize=(10,10))
sns.heatmap(corr_matrix,cmap='coolwarm',annot=True)
plt.title('Correlation distribution plot')
plt.show()

X=data[['AQI', 'PM10', 'O3', 'SO2', 'NO2', 'CO', 'V13305',
       'V10004_700', 'V11291_700', 'V12001_700', 'V13003_700']]
y=data['PM2.5']
rf=RandomForestRegressor(n_estimators=100,random_state=42)
rf.fit(X,y)


feature_importance=rf.feature_importances_
feature_names=X.columns
sorted_idx=feature_importance.argsort()

plt.barh(range(len(sorted_idx)),feature_importance[sorted_idx])
plt.yticks(range(len(sorted_idx)),feature_names[sorted_idx])
plt.xlabel('Importance')
plt.title('Feature Importance')
plt.show()

```

```python
import pandas as pd
import numpy as np

df=pd.read_excel('clean_data/PM2.5.xlsx')
df['date'] = pd.to_datetime(df['date'])
df=df.set_index('date')
df=df[['PM2.5']].dropna()
df.index = pd.date_range(start='2015-01-01',periods=len(df),freq='D')

train_data=df.loc['2015-01-01':'2023-01-01']
test_data = df.loc['2023-01-02':]


from sklearn.preprocessing import MinMaxScaler
# 
scaler = MinMaxScaler(feature_range=(0,1))
# 01
train_data_scaled = scaler.fit_transform(train_data)
test_data_scaled = scaler.transform(test_data)


def create_database(X,y,time_steps=50):
	Xs=[]
	ys=[]
	for i in range(len(X) - time_steps):
		v=X[i:i+time_steps]
		Xs.append(v)
		ys.append(y[i+time_steps])
	return np.array(Xs),np.array(ys)

time_steps=50
X_train,y_train = create_database(train_data_scaled,train_data_scaled[:],time_steps=time_steps)
X_test,y_test = create_database(test_data_scaled,test_data_scaled[:],time_steps=time_steps)
# 
kk = scaler.inverse_transform(test_data_scaled)
kk = kk[:]

from keras.models import Sequential
from keras.layers import LSTM,Dense

model = Sequential()
model.add(LSTM(units=64,input_shape=(X_train.shape[1],X_train.shape[2])))
model.add(Dense(units=8))
model.compile(loss='mean_squared_error',optimizer='adam')


model.fit(X_train,y_train,epochs=50,batch_size=64,validation_split=0.1,verbose=1)

y_pred = model.predict(X_test)
print(y_pred)
y_pred_inv = scaler.inverse_transform(y_test)

yy_pred = y_pred_inv[:]
```

```python
import numpy as np
from keras.models import Sequential
from keras.layers import LSTM, Dense
from datetime import datetime

# 
dates = ['2022-01-01', '2022-01-02', '2022-01-03', '2022-01-04', '2022-01-05']
data = np.array([0, 1, 2, 3, 4])

# 
dates_numeric = [datetime.strptime(date, '%Y-%m-%d').day for date in dates]

# 
X = np.column_stack((dates_numeric, data))

# LSTM
model = Sequential()
model.add(LSTM(50, activation='relu', input_shape=(X.shape[1], 1)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')

# 
train_size = int(len(data) * 0.7)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = data[:train_size], data[train_size:]
model.fit(X_train.reshape((X_train.shape[0], X_train.shape[1], 1)), y_train, epochs=100, batch_size=1, verbose=2)

# 
mse = model.evaluate(X_test.reshape((X_test.shape[0], X_test.shape[1], 1)), y_test)
print('Mean Squared Error:', mse)

# 
future_date = '2022-01-06'
future_date_numeric = datetime.strptime(future_date, '%Y-%m-%d').day
future_input = np.array([[future_date_numeric, 0]])

# 
prediction = model.predict(future_input.reshape((1, future_input.shape[1], 1)))
print('Prediction for', future_date, ':', prediction)
```

```python
# 
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import LSTM, Dense

# 
df = pd.read_excel('clean_data/PM2.5.xlsx')
df['date'] = pd.to_datetime(df['date'])
df=df.set_index('date')
df.index = pd.date_range(start='2015-01-01',periods=len(df),freq='D')
# 
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(df[['PM2.5', 'PM10', 'AQI', 'CO', 'V12001_700', 'NO2', 'O3']])

def create_dataset(data, time_steps=1):
    X, y = [], []
    for i in range(len(data) - time_steps):
        X.append(data[i:i+time_steps, :])
        y.append(data[i+time_steps])  # PM2.5
    return np.array(X), np.array(y)

time_steps = 50
X, y = create_dataset(scaled_data, time_steps)

# 
train_size = int(len(X) * 0.99)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]
y_test_original = df['PM2.5'].iloc[train_size+time_steps:]

# LSTM
model = Sequential()
model.add(LSTM(units=64, input_shape=(X_train.shape[1], X_train.shape[2])))
model.add(Dense(units=7))
model.compile(loss='mean_squared_error', optimizer='adam')

# 
model.fit(X_train, y_train, epochs=50, batch_size=64, validation_split=0.1, verbose=1)

# 
y_pred_normalized = model.predict(X_test)

# 
y_pred = scaler.inverse_transform(y_pred_normalized)[:,0]
# 
import matplotlib.pyplot as plt

plt.plot(y_test_original.index, y_test_original, label='Actual PM2.5')
plt.plot(y_test_original.index, y_pred, label='Predicted PM2.5')
plt.legend()
plt.show()
```

```python
y_pred_normalized = model.predict(X_test)
new_data = np.vstack((scaled_data[train_size:],y_pred_normalized))
last_data = new_data[-time_steps:]
```

```python
predictions = []
for i in range(12):
    y_pred_next=model.predict(np.array([last_data[-time_steps:]]))
    y_pred_next1 = scaler.inverse_transform(y_pred_next)
    predictions.append(y_pred_next1[:,0])
    last_data = np.vstack((last_data,y_pred_next))
```
